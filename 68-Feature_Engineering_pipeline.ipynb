```python
#There are ordinal data in the house.csv

def ordinal_category(col):

    """""""""""""""""
    This function detects ordinal data and assigns them values between 0 - 5.
    
    input -> string/ object
    output -> integer

    """""""""""""""""
    
    if col.isin(["Fa", "TA"," Gd", "Ex"]).any():
        return col.replace({"Fa":2, "TA" : 3, "Gd" : 4, "Ex" : 5, "Po" : 1, "None" : 0})
    else:
        return col
df_new[cat_cols] = df_new[cat_cols].apply(ordinal_category)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeRegressor


# Handles categorical features.
# Fill missing values with the most frequent category.
# convert categorical features into numerical one-hot encoded features.
preprocess_cat = Pipeline([
    ("impute", SimpleImputer(strategy="most_frequent")),
    ("encode", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])


# Handles numerical features.
# Fill missing values with the mean.
# tandardize the numerical features.
preprocess_num = Pipeline([
    ("impute", SimpleImputer(strategy="mean")),
    ("scale", StandardScaler())
])

# Merge the numerical and categorical features.
preprocess = ColumnTransformer([
    ("preprocess_cat", preprocess_cat, cat_cols),
    ("preprocess_num", preprocess_num, num_cols)
])


# Applies the combined preprocessing step.
# Trains a Decision Tree Regressor.
# Selects the top 100 features
ml_pipe = Pipeline([
    ("preprocess", preprocess),
    ("selection", SelectKBest(f_regression, k = 100)),
    ("model", DecisionTreeRegressor(max_depth = 5))
])


#Trains the entire pipeline on the training data.
ml_pipe.fit(X_train , y_train)
ypred = ml_pipe.predict(X_test)
scores = cross_val_score(ml_pipe, X_train, y_train, cv = 10)


results.loc["simple imputer and pipeline"] = [scores.mean(), scores.std(), r2_score(y_test, ypred)]
results

print("Train :", r2_score(y_train, ml_pipe.predict(X_train)))
print("Val :", scores.mean())
print("Test :", r2_score(y_test, ypred))
```
